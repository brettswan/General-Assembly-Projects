##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide.
```python
for j in cleanedTrackRanks:
    tempWeekChange = []
#     print j
#     print len(cleanedTrackRanks[j])
    for i in range(len(cleanedTrackRanks[j])-1):
        tempWeekChange.append(cleanedTrackRanks[j][i+1] - cleanedTrackRanks[j][i])
        
#         print (cleanedTrackRanks[j][i+1] - cleanedTrackRanks[j][i])
#     print tempWeekChange
    cleanedTrackRanks[j].append(tempWeekChange)
#     print tempWeekChange
cleanedTrackRanks
```

```python
RatingCSV = {}
RatingChangeCSV = {}
for i in cleanedTrackRanks:
    Rating = cleanedTrackRanks[i][:-1]
    RatingChange = cleanedTrackRanks[i][-1]
    ExportCSV[i] =  {'Rating' : Rating, 'Rating Change' : RatingChange}
    RatingCSV[i] = Rating
    RatingChangeCSV[i] = RatingChangeCSV
print ExportCSV
#     RatingChangeCSV = RatingChange

# for k,v in RatingCSV.items():
#     print k
#     print v
#     print '<---------------------->'
```

```python
pd.DataFrame(ExportCSV, columns = ['Rating'])
```

```python
Top10 = {}
for i in ExportCSV:
    ratingSet = ExportCSV[i]['Rating']
    counter = 0
    for j in ratingSet:
        if j <= 10:
            counter +=1
    if counter > 5:
        #print i, counter
        Top10[i] = counter
print Top10
```

```python
Top10DF = pd.DataFrame(Top10.items(), columns = ['Track', 'Weeks in Top 10']).sort('Weeks in Top 10', ascending = False)
Top10DF.to_csv('Top10Tracks.csv')
```
