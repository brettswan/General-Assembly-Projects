```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('http://pbpython.com/extras/sample-salesv2.csv')
df
grouped = df.groupby(['name']).mean().sort(['ext price'], ascending = False)
del grouped['account number']
grouped
```

```python
# Create a stacked bar plot of Rep and Price
# Create a stacked bar plot of Rep and Quantity

grouped.plot(kind = 'bar', stacked = False, x = grouped.index, y = 'ext price')
plt.show()
```

```python

grouped.plot(kind = 'bar', stacked = True, x = grouped.index, y = grouped.columns)
plt.show()
```
